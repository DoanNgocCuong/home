```python
import random
import numpy as np
import matplotlib.pyplot as plt

# predict function
def predict(x, b):
    return x + b

def compute_derivative(y_hat, y):
    return 2*(y_hat-y)

def update(b, lr, db):
    return b - lr*db

def show_blackbox(b):
    x_min = -2
    x_max = 10
    x_values = np.linspace(x_min, x_max, 300)
    
    predict_fx = predict(x_values, b)
    
    # Visualization Stuff
    fig, ax = plt.subplots()
    ax.set_xlim([-2, 10])
    ax.set_ylim([-10, 10])
    ax.scatter([4], [6], zorder=4)
    line1, = ax.plot(x_values, predict_fx, 'g')
    ax.grid(True)


# given data 
x = 4
y = 6

b = -5.0
lr = 0.01
```

```python
# current b
show_blackbox(b)
```

```python
# optimize b

for _ in range(200):
    y_hat = predict(x, b)
    db = compute_derivative(y_hat, y)
    b = update(b, lr, db)
```

```python
# current b
show_blackbox(b)
```
