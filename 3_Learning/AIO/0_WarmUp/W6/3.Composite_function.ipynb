### Composite function

```python
'''
f(x) = 2x - 1
g(f) = (f-3)^2

compute the derivative of g(f(x))
''' 
```

```python
import numpy as np
import matplotlib.pyplot as plt

# f function
def fx(x):
    return 2*x - 1

def df_dx(x):
    return 2

# plot f_func
x_list = np.arange(-10, 10, 0.1).tolist()
fx_list = [fx(x) for x in x_list]
df_dx_list = [df_dx(x) for x in x_list]

plt.plot(x_list, fx_list)
plt.plot(x_list, df_dx_list)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# g function
def gf(f):
    return (f-3)**2

def dg_df(f):
    return 2*(f-3)

# plot g_func
f_list = np.arange(-4, 10, 0.1).tolist()
gf_list = [gf(f) for f in f_list]
dg_df_list = [dg_df(f) for f in f_list]

plt.plot(f_list, gf_list)
plt.plot(f_list, dg_df_list)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def fx(x):
    return 2*x - 1

def df_dx(x):
    return 2

def gf(f):
    return (f-3)**2

def dg_df(f):
    return 2*(f-3)

# chain rule
x_list = np.arange(-1, 5, 0.1).tolist()

# forward
fx_list = [fx(x) for x in x_list]
gf_list = [gf(fx) for fx in fx_list]

# backward
df_dx_list = [df_dx(x) for x in x_list]
dg_df_list = [dg_df(fx) for fx in fx_list]
dg_dx_list = [df_dx*dg_df for df_dx, dg_df in zip(df_dx_list, dg_df_list)]

plt.plot(x_list, gf_list)
plt.plot(x_list, dg_dx_list)
plt.show()
```

### Optimization

```python
'''
f(x) = 2x - 1
g(f) = (f-3)^2
''' 
def fx(x):
    return 2*x - 1

def df_dx(x):
    return 2

def gf(f):
    return (f-3)**2

def dg_df(f):
    return 2*(f-3)

# compute g_x
def forward(x):
    fx_value = fx(x)
    gf_value = gf(fx_value)
    return gf_value

# compute dg_x
def backward(x, fx_value):
    df_dx_value = df_dx(x)
    dg_df_value = dg_df(df_dx_value)
    dg_dx_value = df_dx_value*dg_df_value
    return dg_dx_value
```

```python
# minimize the composite function g(f(x))

import random

num_steps = 20
lr = 0.1

# 1. set x randomly
x = random.randint(-100, 100)
print(x)

for _ in range(num_steps):
    # 1. compute derivative
    fx_value = fx(x)
    dg_dx_value = backward(x, fx_value)
            
    # 2 update
    x = x - lr*dg_dx_value
    print(x, '   ', forward(x))
```
