````python
def quadratic_equation(a, b, c):
    """
    Giải Phương Trình Bậc 2
    ax^2 + bx + c = 0
    """
    if a == 0:
        if b == 0:
            if c == 0:
                return "Phương Trình Vô Số Nghiệm"
            else: 
                return "Phương Trình Vô Nghiệm"
        else:
            return -c / b
    else: 
        delta = b**2 - 4*a*c
        if delta < 0:
            return "Phương Trình Vô Nghiệm"
        elif delta == 0:
            return -b / (2*a)
        else:
            # return (-b + sqrt(delta)) / (2*a), (-b - sqrt(delta)) / (2*a)
            return (-b + delta ** 0.5) / (2*a), (-b - delta ** 0.5) / (2*a)

````

````python
def solve_linear_equation(b, c):
    """
    Giải Phương Trình Bậc 1
    y = bx + c = 0
    """
    if b == 0:
        if c == 0:
            return "Phương Trình Vô Số Nghiệm"
        else: 
            return "Phương Trình Vô Nghiệm"
    return -c / b

def solve_quadratic_equation(a, b, c):
    """
    Giải Phương Trình Bậc 2
    ax^2 + bx + c = 0
    """
    if a == 0:
        return solve_linear_equation(b, c)
    else:
        delta = b**2 - 4*a*c
        if delta < 0:
            return "Phương Trình Vô Nghiệm"
        elif delta == 0:
            return -b / (2*a)
        else:
            return (-b + delta ** 0.5) / (2*a), (-b - delta ** 0.5) / (2*a)

````

### Cách tiếp cận của Newton với gradient descent

```
số lần huấn luyện ít hơn, có phải do bước nhảy lớn hơn không ạ,
```
---
- Gradient descent: thì x được cập nhật từ từ theo learning rate. 
- Newton: thì x với bước nhảy lớn, có thể nhảy qua nhảy lại. 


### **So sánh Gradient Descent và Phương pháp Newton qua bảng:**

| **Tiêu chí**               | **Gradient Descent**                                                                 | **Phương pháp Newton (Newton-Raphson)**                                           |
|---------------------------|--------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------|
| **Cách cập nhật**          | Cập nhật từ từ theo đạo hàm, dựa trên learning rate:                                 | Cập nhật bằng cách dùng cả đạo hàm và đạo hàm bậc hai:                            |
|                           | \[ x_{n+1} = x_n - \alpha \cdot f'(x_n) \]                                           | \[ x_{n+1} = x_n - \frac{f'(x_n)}{f''(x_n)} \]                                    |
| **Tốc độ hội tụ**          | Thường chậm hơn vì dựa vào bước nhảy cố định (learning rate).                       | Thường nhanh hơn, đặc biệt khi điểm bắt đầu gần nghiệm (hội tụ bậc hai).          |
| **Bước nhảy (step size)**  | Phụ thuộc vào learning rate \( \alpha \). Bước nhảy nhỏ, từ từ giảm dần.            | Bước nhảy lớn hơn, có thể nhảy qua nhảy lại tùy thuộc vào độ cong (curvature).    |
| **Độ ổn định**             | Tương đối ổn định khi chọn learning rate nhỏ.                                       | Dễ không ổn định nếu đạo hàm bậc hai gần bằng 0 (nhảy quá xa hoặc không hội tụ).   |
| **Khi nào hiệu quả hơn**   | Phù hợp khi tối ưu hàm có dạng phức tạp hoặc không có đạo hàm bậc hai.              | Hiệu quả hơn với các bài toán có đạo hàm bậc hai và điểm bắt đầu gần nghiệm.      |
| **Tính toán mỗi bước**     | Dễ dàng tính toán vì chỉ cần đạo hàm bậc một.                                       | Phức tạp hơn vì cần cả đạo hàm và đạo hàm bậc hai.                                |
| **Khả năng nhảy ra local minimum** | Khó nhảy ra khỏi các điểm local minimum do bước nhảy cố định.               | Có khả năng vượt qua local minimum nhờ bước nhảy lớn và hướng dẫn từ đạo hàm bậc hai. |
| **Ứng dụng thực tế**       | Sử dụng rộng rãi trong các thuật toán tối ưu lớn (như trong Machine Learning).      | Thường dùng khi cần tìm nghiệm chính xác cho các phương trình phi tuyến.         |

### **Kết luận:**
- **Gradient Descent** ổn định hơn, dễ dùng trong các bài toán tối ưu lớn.
- **Newton** hội tụ nhanh hơn nhưng có thể không ổn định khi xa nghiệm hoặc khi \( f''(x) \) gần bằng 0.

- Chiếu 1000 chiều qua 1 Random Projections => Để chiếu qua 1 chiều không gian mới, để ở chiều không gian đó mình sẽ tính distance này 

````python
# Example of Wasserstein Distance

# Create 
P = [0.1, 0.2, 0.3, 0.4]
Q = [0.5, 0.4, 0.3, 0.2]

P_cum = [0] * len(P)
Q_cum = [0] * len(Q)

for i in range(len(P)):
    P_cum[i] = P[i] + P[i-1]
    print(P_cum[i])
    Q_cum[i] = Q[i] + Q[i-1]
    print(Q_cum[i])

print(P_cum)
print(Q_cum)
````

### Giải thích các thành phần:
1. **S**: Ảnh nguồn phong cách (ví dụ tranh Van Gogh).
2. **X**: Ảnh đầu vào ngẫu nhiên (ban đầu có thể giống hoặc khác ảnh nội dung).
3. **C**: Ảnh nội dung (ảnh thực tế hoặc chân dung mà bạn muốn giữ nội dung chính).

4. **VGG**: Mạng CNN dùng để trích xuất đặc trưng (feature extraction). Thường là mạng VGG-19, vì nó mạnh trong nhận diện đặc trưng hình ảnh.

5. **Adaptive Sampling / Uniform Sampling**: Các kỹ thuật lấy mẫu khác nhau để xác định các đặc trưng quan trọng từ phong cách và nội dung.

6. **Random Projections**: Dự đoán ngẫu nhiên và ánh xạ đặc trưng vào không gian chung.

7. **Loss Functions**:
   - $\mathcal{L}_{style}$: Tổn thất về phong cách, giúp tối ưu hóa để ảnh đầu ra giữ được nét vẽ, màu sắc, và họa tiết của ảnh phong cách (thường dựa trên ma trận Gram).
   - $\mathcal{L}_{content}$: Tổn thất về nội dung, giúp đảm bảo bố cục và hình dạng tổng thể của ảnh nội dung được giữ lại.
