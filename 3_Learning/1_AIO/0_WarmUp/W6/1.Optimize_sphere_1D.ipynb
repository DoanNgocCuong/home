```python
# sphere function
def func(x):
    return x**2

def derivative(x):     
    return 2*x
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-100, 100, 0.1)
y = func(x)

plt.plot(x, y)
plt.show()
```

```python
import random

num_steps = 10
step = 1
lr = 0.1

# 1. set x randomly
x = random.randint(-100, 100)
print(x)

for _ in range(num_steps):
    # compute derivative
    dx = derivative(x)
        
    # update
    x = x - lr*step
    print(x, func(x))
```
