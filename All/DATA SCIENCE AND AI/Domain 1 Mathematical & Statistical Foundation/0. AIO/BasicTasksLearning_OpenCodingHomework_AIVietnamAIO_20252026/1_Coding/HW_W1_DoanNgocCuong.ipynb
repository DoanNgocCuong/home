#### Bài tập: Tính lịch Can Chi

````python
def calculate_can_chi_calendar(year):
    year = int(year)
    year_can = year % 10
    year_chi = year % 12
    
    # Canh Tân Nhâm Quý Giáp Ất Bính Đinh Mậu Kỷ
    can_chi_calendar = {
        0: 'Canh',
        1: 'Tân',
        2: 'Nhâm',
        3: 'Quý',
        4: 'Giáp',
        5: 'Ất',
        6: 'Bính',
        7: 'Đinh',
        8: 'Mậu',
        9: 'Kỷ',
    }
    # Thân Dậu Tuất Hợi Tý Sửu Dần Mẹo Thìn Tỵ Ngọ Mùi
    chi_calendar = {
        0: 'Thân',
        1: 'Dậu',
        2: 'Tuất',
        3: 'Hợi',
        4: 'Tý',
        5: 'Sửu',
        6: 'Dần',
        7: 'Mão',
        8: 'Thìn',
        9: 'Tỵ',
        10: 'Ngọ',
        11: 'Mùi',
    }
    result = can_chi_calendar[year_can] + " " + chi_calendar[year_chi]
    return result


# - Test case 1: calculate_can_chi_calendar(2025) → Ất Tỵ
# - Test case 2: calculate_can_chi_calendar(2024) → Giáp Thìn
# - Test case 3: calculate_can_chi_calendar(1997) → Đinh S

print(calculate_can_chi_calendar(2025))
print(calculate_can_chi_calendar(2024))
print(calculate_can_chi_calendar(1997))
````

## 2 Hàm kích hoạt- Activate function

### 2.2.1  ReLU (Rectified Linear Unit)
```
ReLU(x) = x if x>0
          0 if x≤0
```



````python
def ReLU(x):
    if x > 0:
        return x
    else:
        return 0

print(ReLU(3))
print(ReLU(-2))
print(ReLU(0))
````

````python
def ReLU_2(x):
    return max(0, x)

print(ReLU_2(3))
print(ReLU_2(-2))
print(ReLU_2(0))
````

````python
def ReLU_3(x):
    return x if x > 0 else 0

print(ReLU_3(3))
print(ReLU_3(-2))
print(ReLU_3(0))

````

### 2.2.2 LeakyReLU
```
LeakyReLU(x) = x if x>0
              Alpha*x if x≤0
```



````python
def leaky_relu(x, alpha=0.01):
    return x if x > 0 else alpha * x

print(leaky_relu(3))
print(leaky_relu(-2))
print(leaky_relu(0))
````

### 2.2.3 Sigmoid
```
Sigmoid(x) = 1 / (1 + e^(-x))
```



````python
import math
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

print(sigmoid(0))
print(sigmoid(2))
print(sigmoid(-2))
````

### 2.2.4 Tanh
```
Tanh(x) = (e^x - e^(-x)) / (e^x + e^(-x))
```


````python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

print(tanh(0))
print(tanh(2))
print(tanh(-2))
````

### 2.2.5  ELU (Exponential Linear Unit)
```
ELU(x) = x if x > 0
         α(ex −1) if x ≤ 0
```



````python
def elu(x, alpha=1.0):
    return x if x > 0 else alpha * (math.exp(x) - 1)

print(elu(3))
print(elu(-1))
print(elu(0))
````

### 2.2.6 Sign
```
Sign(x) = 1 if x > 0
          0 if x = 0
          -1 if x < 0
```

````python
def sign(x):
    return 1 if x > 0 else -1 if x < 0 else 0

# Ví dụ sử dụng:
print(sign(10))   # In ra 1
print(sign(-5))   # In ra -1
print(sign(0))    # In ra 0
````
