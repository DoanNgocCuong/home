### Các bước thực hiện:

1. **Cập nhật dependencies**:
```bash
cd frontend
npm install --save-dev @babel/plugin-proposal-private-property-in-object
```

2. **Xóa node_modules và cài lại**:
```bash
rm -rf node_modules
npm install
```

3. **Khởi động lại ứng dụng**:
```bash
npm start
```
