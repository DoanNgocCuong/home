FileUploader.js là một component con (child component) được sử dụng trong FileUpload.js. Đây là mối quan hệ parent-child, cụ thể:

1. **FileUpload.js (Parent/Cha)**
- Chứa logic xử lý (state và functions)
- Import và sử dụng FileUploader component
- Truyền props xuống cho FileUploader
```javascript
// FileUpload.js
import FileUploader from './components/FileUploader';

function FileUpload() {
    const [file, setFile] = useState(null);
    const [loading, setLoading] = useState(false);
    
    const handleFileChange = (e) => {
        setFile(e.target.files[0]);
    };

    const handleUpload = async () => {
        // xử lý upload...
    };

    return (
        <FileUploader 
            onFileChange={handleFileChange}  // truyền function xuống
            onUpload={handleUpload}          // truyền function xuống
            loading={loading}                 // truyền state xuống
            file={file}                      // truyền state xuống
        />
    );
}
```

2. **FileUploader.js (Child/Con)**
- Chỉ chứa UI (giao diện)
- Nhận props từ component cha
- Không có state hay logic xử lý
```javascript
// FileUploader.js
function FileUploader({ onFileChange, onUpload, loading, file }) {
    return (
        <div className="upload-section">
            <input 
                type="file" 
                onChange={onFileChange}  // sử dụng function từ cha
                accept=".xlsx,.xls"
                disabled={loading}       // sử dụng state từ cha
            />
            <button 
                onClick={onUpload}       // sử dụng function từ cha
                disabled={loading || !file}  // sử dụng state từ cha
            >
                {loading ? 'Uploading...' : 'Upload'}
            </button>
        </div>
    );
}
```

Lợi ích của cách tổ chức này:
1. **Tách biệt trách nhiệm**:
   - FileUpload.js: Xử lý logic
   - FileUploader.js: Hiển thị UI

2. **Dễ tái sử dụng**:
   - FileUploader là component thuần UI
   - Có thể dùng lại với các logic khác nhau

3. **Dễ test**:
   - Có thể test UI và logic riêng biệt
   - Dễ mock props để test FileUploader

4. **Dễ bảo trì**:
   - Muốn sửa logic: sửa FileUpload.js
   - Muốn sửa giao diện: sửa FileUploader.js

5. **Props rõ ràng**:
   - FileUploader chỉ nhận các props cần thiết
   - Interface giữa 2 component rõ ràng

Đây là một ví dụ tốt về việc áp dụng nguyên tắc "separation of concerns" trong React, tách biệt phần logic (container component) và phần UI (presentational component).

Ta có 2 cách xử lý:

1. Cách 1: Thêm prop `filename` vào `FileUploader.js`:

```javascript
// FileUploader.js
function FileUploader({ onFileChange, onUpload, loading, file, filename }) {
    return (
        <div className="upload-section">
            <input 
                type="file" 
                onChange={onFileChange}
                accept=".xlsx,.xls"
                disabled={loading}
            />
            <button 
                onClick={onUpload}
                disabled={loading || !file}
                className="upload-button"
            >
                {loading ? 'Uploading...' : 'Upload'}
            </button>
            
            {filename && (
                <div className="success-message">
                    File {filename} uploaded successfully!
                </div>
            )}
        </div>
    );
}

// Upload.js sẽ truyền filename xuống:
<FileUploader 
    onFileChange={handleFileChange}
    onUpload={handleUpload}
    loading={loading}
    file={file}
    filename={filename}  // thêm prop này
/>
```

2. Cách 2: Tạo một component thông báo riêng:

```javascript
// UploadMessage.js
function UploadMessage({ filename, error }) {
    return (
        <div className="upload-messages">
            {filename && (
                <div className="success-message">
                    File {filename} uploaded successfully!
                </div>
            )}
            {error && (
                <div className="error-message">{error}</div>
            )}
        </div>
    );
}

// Upload.js sẽ dùng cả 2 components:
return (
    <div className="container">
        <FileUploader 
            onFileChange={handleFileChange}
            onUpload={handleUpload}
            loading={loading}
            file={file}
        />
        <UploadMessage 
            filename={filename}
            error={error}
        />
        ...
    </div>
);
```

Tôi khuyến nghị dùng cách 2 vì:
1. Tách biệt rõ ràng giữa phần UI upload và phần thông báo
