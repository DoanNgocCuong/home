Dựa vào đường dẫn, ta có thể lấy thông tin chi tiết của conversation bằng cách:

```bash
# 1. Lấy log từ workflows 
curl --location 'https://studio.bavaan.ai/v1/workflows/logs' \
--header 'Authorization: Bearer app-HhMMVMgdZyAOPWv3gKnRJgop'

# 2. Từ log lấy workflow_run_id, sau đó lấy chi tiết:
curl --location 'https://studio.bavaan.ai/v1/workflows/run/{workflow_run_id}' \
--header 'Authorization: Bearer app-HhMMVMgdZyAOPWv3gKnRJgop'
```

Thay {workflow_run_id} bằng ID từ response của API logs.

Response sẽ chứa:
- ID của workflow run
- Status và thời gian thực thi
- Inputs/outputs của các node 
- Execution metadata (lịch sử chat)
