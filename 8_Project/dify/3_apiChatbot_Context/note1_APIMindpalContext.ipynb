Mỗi response sẽ chứa thông tin chi tiết về trạng thái và dữ liệu của node đó trong conversation.

Bạn có thể thêm hàm helper vào `mindpal.ts` để đơn giản hóa việc này:

```typescript
export const mindpalFetchAllNodesOfConversation = async (
  appId: string,
  conversationId: string
) => {
  // 1. Fetch workflow to get node IDs
  const workflow = await consoleGet(`apps/${appId}/workflows/publish`);
  const nodeIds = workflow.graph.nodes.map(node => node.id);
  
  // 2. Fetch each node execution
  const nodeExecutions = await Promise.all(
    nodeIds.map(nodeId => 
      mindpalFetchNodeExecutionDetailsOfConversation(
        appId,
        conversationId, 
        nodeId
      )
    )
  );

  return nodeExecutions;
}
```

Hàm này sẽ tự động:
1. Lấy workflow để có danh sách node IDs
2. Gọi API lấy chi tiết cho từng node
