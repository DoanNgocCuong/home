## Chưa hiểu lắm, khác gì nhau đâu nếu không gạch dưới, _ 1 gạch và __ 2 gạch 

**Sự khác biệt giữa không gạch dưới, một gạch dưới `_` và hai gạch dưới `__` trong Python:**

Đây là một trong những khía cạnh quan trọng về quy ước đặt tên biến trong Python, ảnh hưởng đến tính đóng gói (encapsulation) và phạm vi truy cập của thuộc tính hoặc phương thức trong lớp. Hãy cùng phân biệt rõ:

---

**1. Không gạch dưới (ví dụ: `age`):**

Khi bạn đặt tên biến hoặc phương thức mà không có gạch dưới nào ở đầu, ví dụ `age`, thì:

- Biến hoặc phương thức này là **public** (công khai).
- Có thể truy cập, sửa đổi hoặc gọi từ bất kỳ đâu, cả bên trong lẫn ngoài lớp.

**Ví dụ:**

```python
class Cat:
    age = 0
    def describe(self):
        print(self.age)

cat = Cat()
cat.describe()     # In ra 0
print(cat.age)     # Truy cập trực tiếp, vẫn in ra 0
```

---

**2. Một gạch dưới đầu (ví dụ: `_age`):**

Khi bạn đặt tên biến với một dấu gạch dưới ở đầu, ví dụ `_age`, thì:

- Theo **quy ước** của Python (PEP 8), đây là thuộc tính **protected** (bán riêng tư).
- Ý nghĩa là: "Đây là biến nội bộ, không nên truy cập trực tiếp từ bên ngoài lớp (hoặc module), nhưng Python không ngăn cản".
- Bạn **vẫn có thể truy cập** biến này từ bên ngoài, Python chỉ nhắc nhở qua quy ước cho lập trình viên.

**Ví dụ:**

```python
class Cat:
    _age = 0
    def describe(self):
        print(self._age)

cat = Cat()
cat.describe()     # In ra 0
print(cat._age)    # Truy cập được, nhưng không khuyến khích
```

---

**3. Hai gạch dưới đầu (ví dụ: `__age`):**

Khi bạn đặt tên biến với hai dấu gạch dưới ở đầu, ví dụ `__age`, thì:

- Python sẽ thực hiện **name mangling**: tự động đổi tên biến thành `_TênLớp__TênBiến` để tránh bị ghi đè hoặc truy cập ngoài ý muốn, đặc biệt trong trường hợp kế thừa.
- Biến này được coi là **private** (riêng tư thực sự), chỉ nên truy cập trong nội bộ lớp.
- Nếu bạn cố gắng truy cập trực tiếp từ bên ngoài (ví dụ: `cat.__age`), bạn sẽ gặp lỗi.

**Ví dụ:**

```python
class Cat:
    __age = 0
    def describe(self):
        print(self.__age)

cat = Cat()
cat.describe()     # In ra 0
print(cat.__age)   # Lỗi: AttributeError
print(cat._Cat__age) # Truy cập được, tên đã bị đổi
```

---

**Tóm tắt sự khác biệt:**

- **Không gạch dưới (`age`)**: Public, truy cập thoải mái ở mọi nơi.
- **Một gạch dưới (`_age`)**: Protected (quy ước), nên giới hạn truy cập bên ngoài, nhưng vẫn truy cập được.
- **Hai gạch dưới (`__age`)**: Private (name mangling), hạn chế truy cập bên ngoài, tên biến bị đổi thành `_Cat__age` (nếu lớp là Cat).

---

**Bảng tóm tắt:**

| Loại           | Truy cập ngoài lớp | Được khuyên dùng khi      | Tên thực tế trong Python object |
|----------------|-------------------|---------------------------|---------------------------------|
| `age`          | Có                | Biến công khai (public)   | `age`                           |
| `_age`         | Có                | Biến nội bộ/protected     | `_age`                          |
| `__age`        | Không trực tiếp   | Biến private thực sự      | `_Cat__age` (nếu lớp là Cat)    |

---

**Kết luận:**

- **Không gạch dưới**: dùng cho thuộc tính hoặc phương thức công khai.
- **Một gạch dưới**: dùng khi muốn nhắc nhở đây là thuộc tính nội bộ, không nên đụng vào từ bên ngoài.
- **Hai gạch dưới**: dùng khi thật sự muốn bảo vệ thuộc tính khỏi bị truy cập hoặc ghi đè ngoài ý muốn, đặc biệt trong các dự án lớn hoặc có kế thừa lớp.

## Khi nào cần self. ; khi nào không?

**Khi nào cần dùng `self.` và khi nào không trong Python:**

Việc sử dụng `self` trong Python liên quan trực tiếp đến phạm vi của biến hoặc phương thức mà bạn muốn truy cập trong lớp. Đây là điểm rất quan trọng để phân biệt giữa biến/tham số cục bộ và biến/thành viên của đối tượng. Dưới đây là giải thích rõ ràng và cụ thể:

---

**1. Khi nào cần dùng `self.`:**

Bạn cần dùng `self.` khi muốn truy cập đến **thuộc tính (biến) hoặc phương thức của đối tượng (instance)**, tức là các biến hoặc hàm được gắn với từng đối tượng cụ thể của lớp.

- Khi bạn muốn đọc hoặc gán giá trị cho một biến thành viên của đối tượng (instance variable).
- Khi bạn muốn gọi một phương thức khác của cùng đối tượng.
- Khi bạn muốn cập nhật hoặc sử dụng dữ liệu lưu trữ trong từng đối tượng riêng biệt.

**Ví dụ:**

```python
class Cat:
    def __init__(self, age):
        self.age = age   # self.age là biến thành viên

    def describe(self):
        print(self.age)  # Dùng self.age để truy cập tuổi của đối tượng hiện tại

cat1 = Cat(3)
cat1.describe()    # In ra 3
```

Ở đây, `self.age` chính là thuộc tính của từng đối tượng `cat1`, `cat2`,... Nếu không có `self.`, bạn chỉ tạo ra biến cục bộ trong hàm, không lưu được giá trị ngoài hàm.

---

**2. Khi nào không cần dùng `self.`:**

Bạn **không cần dùng `self.`** khi:

- Truy cập vào **biến cục bộ** hoặc **tham số** được truyền vào hàm (không phải là thuộc tính của đối tượng).
- Viết các hàm hoặc biến chỉ dùng trong phạm vi của phương thức đó mà không cần lưu trữ cho đối tượng.

**Ví dụ:**

```python
class Cat:
    def set_age(self, age):
        age = age + 1    # age ở đây chỉ là biến cục bộ/tham số, không ảnh hưởng tới self.age

    def print_something(self):
        x = 5            # x là biến cục bộ, chỉ dùng trong hàm này
        print(x)
```

Trong ví dụ này, `age` và `x` chỉ tồn tại trong hàm, không lưu lại trong đối tượng. Nếu bạn không dùng `self.`, các giá trị này sẽ biến mất sau khi hàm chạy xong.

---

**3. Điều gì xảy ra nếu nhầm lẫn giữa có và không có `self.`:**

- Nếu bạn gán giá trị cho một biến mà không có `self.`, biến đó chỉ tồn tại trong hàm, không phải là thuộc tính của đối tượng.
- Nếu bạn muốn thuộc tính đó thuộc về đối tượng và truy cập được ở các phương thức khác, bạn **phải** dùng `self.`.

**Ví dụ minh hoạ sự nhầm lẫn:**

```python
class Cat:
    def set_age(self, age):
        self.age = age   # Đúng: Lưu giá trị cho đối tượng

    def set_age_wrong(self, age):
        age = age        # Sai: Chỉ gán lại cho biến cục bộ, không lưu gì cho đối tượng

    def show(self):
        print(self.age)

cat = Cat()
cat.set_age(5)
cat.show()        # In ra 5

cat2 = Cat()
cat2.set_age_wrong(3)
cat2.show()       # Báo lỗi: 'Cat' object has no attribute 'age'
```

---

**Tóm tắt quy tắc dùng `self.`:**

- **Bắt buộc dùng `self.`** khi muốn truy cập hoặc thay đổi thuộc tính/phương thức của đối tượng (instance).
- **Không dùng `self.`** khi chỉ làm việc với biến cục bộ hoặc tham số bên trong phương thức.

---

**Kết luận:**

- Nếu muốn lưu thông tin cho từng đối tượng, hoặc muốn các phương thức khác cũng truy cập được thông tin đó, **luôn phải dùng `self.`**.
- Nếu chỉ cần biến tạm thời trong hàm, **không dùng `self.`**.

````python
class Cat:
    _age = 0
    def describe(self):
        print(self._age)

cat = Cat()
cat.describe()     # In ra 0
print(cat._age)    # Truy cập được, nhưng không khuyến khích
````

### biến lớp và biến instance có phải là 1 không?

**Biến lớp (class variable) và biến instance (instance variable) có phải là 1 không?**

Câu trả lời ngắn gọn là: **Không, chúng không phải là một!**  
Chúng là hai loại biến hoàn toàn khác nhau về vai trò, phạm vi và cách truy cập trong Python.

---

**1. Biến lớp (class variable):**

- Biến lớp là biến được khai báo **trực tiếp trong thân lớp**, bên ngoài mọi phương thức.
- Biến lớp **dùng chung cho tất cả các đối tượng** của lớp đó. Nếu bạn thay đổi biến lớp từ chính lớp hoặc từ một đối tượng mà chưa từng gán biến instance trùng tên, thì tất cả các đối tượng khác cũng nhìn thấy giá trị mới.
- Bạn truy cập biến lớp qua `Cat.age` hoặc `self.age` (nếu chưa có biến instance trùng tên).

**Ví dụ:**
```python
class Cat:
    age = 0  # biến lớp

cat1 = Cat()
cat2 = Cat()

print(cat1.age)  # 0
print(cat2.age)  # 0

Cat.age = 10     # thay đổi biến lớp

print(cat1.age)  # 10
print(cat2.age)  # 10
```

---

**2. Biến instance (instance variable):**

- Biến instance là biến được khai báo **trong phương thức** (thường là trong `__init__`), với cú pháp `self.tenbien = ...`.
- Mỗi đối tượng (instance) **có một bản sao riêng biệt** của biến này. Thay đổi biến instance của một đối tượng không ảnh hưởng đến các đối tượng khác.
- Bạn luôn truy cập biến instance qua `self.tenbien`.

**Ví dụ:**
```python
class Cat:
    def __init__(self, age):
        self.age = age  # biến instance

cat1 = Cat(2)
cat2 = Cat(5)

print(cat1.age)  # 2
print(cat2.age)  # 5
```

---

**3. Sự khác nhau khi truy cập và ghi đè:**

Nếu một đối tượng chưa từng gán giá trị cho biến instance trùng tên với biến lớp, thì khi truy cập thuộc tính đó qua `self.age`, Python sẽ lấy giá trị từ biến lớp.  
Nếu bạn gán `self.age = ...` ở đâu đó, thì từ thời điểm đó trở đi, đối tượng đó sẽ có biến instance riêng và không còn dùng biến lớp nữa.

**Ví dụ minh họa:**
```python
class Cat:
    age = 0

cat1 = Cat()
cat2 = Cat()

cat1.age = 7         # tạo biến instance cho cat1

print(cat1.age)      # 7  (biến instance của cat1)
print(cat2.age)      # 0  (vẫn dùng biến lớp)
print(Cat.age)       # 0

Cat.age = 99         # thay đổi biến lớp

print(cat1.age)      # 7  (không đổi)
print(cat2.age)      # 99 (cat2 dùng biến lớp nên đổi theo)
```

---

**Tóm lại:**

- **Biến lớp:**  
  - Khai báo ngoài mọi phương thức.
  - Dùng chung cho tất cả các đối tượng.
  - Truy cập qua lớp hoặc qua đối tượng (nếu chưa có biến instance trùng tên).

- **Biến instance:**  
  - Khai báo với `self.` trong phương thức (thường là `__init__`).
  - Mỗi đối tượng có một giá trị riêng biệt.
  - Truy cập qua `self.`.

**Chúng không phải là một, mà độc lập với nhau. Khi chưa có biến instance trùng tên, truy cập qua đối tượng sẽ lấy giá trị biến lớp. Khi đã gán biến instance, đối tượng sẽ ưu tiên biến instance đó.**

**Giải thích vì sao đoạn code này vẫn chạy được:**

```python
class Cat:
    def describe(age, this):
        age.__age = 9 
        print(age.__age, this)

cat = Cat()
cat.describe(1)
```

---

**1. Python không ép buộc tên tham số đầu tiên:**

- Khi bạn định nghĩa một phương thức trong lớp, tham số đầu tiên luôn **nhận chính đối tượng được gọi** (ở đây là `cat`) – theo quy ước nên đặt tên là `self` cho dễ đọc và dễ hiểu.
- Tuy nhiên, **Python không bắt buộc** bạn phải đặt tên là `self`. Bạn có thể đặt tên khác, ví dụ `age` như trong code của bạn.  
- Khi gọi `cat.describe(1)`, Python tự động truyền đối tượng `cat` vào tham số đầu tiên (`age`), còn `1` vào tham số thứ hai (`this`).

---

**2. Gán và truy cập thuộc tính động:**

- Dòng `age.__age = 9` thực chất là gán thuộc tính (biến) `__age` cho đối tượng `cat`.  
- Trong Python, bạn có thể thêm thuộc tính mới cho đối tượng bất cứ khi nào, kể cả bên ngoài hàm `__init__`.
- Dù tên biến là `__age`, Python thực hiện *name mangling* nên nó sẽ lưu thành `cat._Cat__age`.

- Dòng `print(age.__age, this)` sẽ lấy thuộc tính vừa gán (`__age`, thực chất là `_Cat__age`) của đối tượng `cat` và in ra tham số thứ hai (`this`), là `1`.

---

**3. Kết quả thực tế:**

Khi chạy, kết quả in ra sẽ là:
```
9 1
```
- Vì Python hiểu:
  - `age` chính là đối tượng `cat`
  - `age.__age` (thực tế là `cat._Cat__age`) vừa được gán bằng 9
  - `this` nhận giá trị 1

---

**4. Tổng kết lý do chạy được:**

- **Python không bắt buộc tên tham số phải là `self`** – chỉ cần đúng vị trí (tham số đầu tiên nhận đối tượng).
- **Có thể tạo thuộc tính mới cho đối tượng bất cứ lúc nào**.
- **Tham số tên gì không quan trọng**, miễn là khi gọi đúng số lượng tham số.
- **Name mangling** chỉ đổi tên thuộc tính trong nội bộ, nhưng không cản trở việc gán và truy cập trong phương thức.

---

**Khuyến nghị về mặt code sạch:**

- Bạn nên **luôn đặt tên tham số đầu tiên là `self`** để code dễ hiểu và tránh nhầm lẫn.
- Việc đặt tên `age` cho đối tượng sẽ gây nhầm lẫn với ý nghĩa thông thường của từ này.

### **Biến lớp và biến instance khác nhau như thế nào?**

# **TỔNG HỢP TOÀN BỘ NỘI DUNG OOP PYTHON VÀ CHECKLIST LỖI THƯỜNG GẶP**

---

## **1. TÓM TẮT TOÀN BỘ NỘI DUNG CHÍNH**

### **Phân biệt các loại tên biến trong Python:**

**Bảng tổng hợp:**

| **Loại** | **Ý nghĩa** | **Truy cập từ bên ngoài** | **Name mangling** | **Mục đích sử dụng** |
|----------|-------------|---------------------------|-------------------|---------------------|
| `variable` | Public (công khai) | Có | Không | Thuộc tính/phương thức công khai |
| `_variable` | Protected (bán riêng tư) | Có (không khuyến khích) | Không | Thuộc tính nội bộ, chỉ dùng trong lớp |
| `__variable` | Private (riêng tư) | Không trực tiếp | Có (`_ClassName__variable`) | Thuộc tính riêng tư, tránh xung đột |

**Chi tiết từng loại:**

- **Không gạch dưới (`age`)**: Thuộc tính công khai, truy cập thoải mái ở mọi nơi
- **Một gạch dưới (`_age`)**: Quy ước protected, Python không ngăn cản nhưng không nên truy cập từ bên ngoài
- **Hai gạch dưới (`__age`)**: Python thực hiện *name mangling*, đổi tên thành `_Cat__age` để bảo vệ thực sự

---

### **Quy tắc sử dụng `self`:**

**Khi nào BẮT BUỘC dùng `self.`:**
- Truy cập hoặc gán giá trị cho thuộc tính của đối tượng (instance variable)
- Gọi phương thức khác của cùng đối tượng
- Lưu trữ dữ liệu cần tồn tại sau khi phương thức kết thúc

**Khi nào KHÔNG cần `self.`:**
- Làm việc với biến cục bộ trong phương thức
- Sử dụng tham số được truyền vào hàm
- Tính toán tạm thời không cần lưu trữ

---

### **Biến lớp vs Biến instance:**

**Biến lớp (Class Variable):**
- Khai báo bên ngoài mọi phương thức
- Dùng chung cho tất cả đối tượng của lớp
- Truy cập qua `ClassName.variable` hoặc `self.variable` (nếu chưa có instance variable cùng tên)

**Biến instance (Instance Variable):**
- Khai báo với `self.variable` trong phương thức (thường trong `__init__`)
- Mỗi đối tượng có một bản sao riêng
- Luôn truy cập qua `self.variable`

**Thứ tự ưu tiên:** Instance variable → Class variable → AttributeError

---

### **Kế thừa và `super()`:**

**Quy tắc quan trọng:**
- Nếu lớp con định nghĩa lại `__init__`, PHẢI gọi `super().__init__()` để kế thừa thuộc tính từ lớp cha
- Thứ tự gọi `super().__init__()` ảnh hưởng đến giá trị cuối cùng của thuộc tính
- Gán sau cùng sẽ là giá trị cuối cùng

### 1.2. SỬ DỤNG `self`

#### **Khi nào cần `self.`:**

        return temp

### 1.3. BIẾN LỚP VS BIẾN INSTANCE

#### **Biến lớp (Class Variable)**
- Khai báo bên ngoài mọi phương thức
- Dùng chung cho tất cả đối tượng
- Truy cập qua `ClassName.variable` hoặc `self.variable`

#### **Biến instance (Instance Variable)**
- Khai báo với `self.` trong phương thức
- Riêng cho từng đối tượng
- Truy cập qua `self.variable`

```python
class Cat:

    def __init__(self, name):

cat1 = Cat("Tom")

print(cat1.name)    # "Tom" - biến instance
print(cat2.name)    # "Jerry" - biến instance
```

## 3. BEST PRACTICES

### ✅ **Quy tắc vàng:**

1. **Luôn dùng `self` cho tham số đầu tiên** của phương thức instance
2. **Gọi `super().__init__()`** khi lớp con có constructor riêng
3. **Dùng `self.`** khi muốn lưu trữ dữ liệu cho đối tượng
4. **Tuân thủ quy ước underscore** cho mức độ truy cập
5. **Kiểm tra thứ tự** khi gọi `super()` và gán thuộc tính

### ✅ **Ví dụ hoàn chỉnh:**

```python
class Animal:
    species_count = 0  # Biến lớp
    
    def __init__(self, name):
        self.name = name  # Biến instance
        Animal.species_count += 1
        self._id = Animal.species_count  # Protected
        self.__secret = "secret_data"  # Private
    
    def get_info(self):
        return f"Name: {self.name}, ID: {self._id}"
    
    def _internal_method(self):  # Protected method
        return self.__secret

class Cat(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # Gọi constructor lớp cha
        self.breed = breed  # Thuộc tính riêng của Cat
    
    def get_full_info(self):
        base_info = self.get_info()  # Gọi phương thức lớp cha
        return f"{base_info}, Breed: {self.breed}"

# Sử dụng
cat = Cat("Tom", "Persian")
print(cat.get_full_info())  # Name: Tom, ID: 1, Breed: Persian
print(Animal.species_count)  # 1
```

## 4. TÓM TẮT NHANH

| Vấn đề | Cần nhớ | Ví dụ |
|--------|---------|-------|
| **Truy cập thuộc tính** | Dùng `self.` | `self.name = value` |
| **Biến private** | Truy cập qua `self.__var` | `print(self.__age)` |
| **Kế thừa** | Gọi `super().__init__()` | `super().__init__(params)` |
| **Biến lớp** | Dùng chung cho tất cả đối tượng | `ClassName.variable` |
| **Biến instance** | Riêng cho từng đối tượng | `self.variable` |
| **Tham số đầu tiên** | Luôn là `self` | `def method(self, ...)` |

**Nhớ:** Khi có lỗi, kiểm tra theo thứ tự: `self.` → `super()` → tham số → quy ước đặt tên!
