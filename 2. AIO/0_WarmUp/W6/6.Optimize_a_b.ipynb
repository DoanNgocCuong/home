```python
import random
import numpy as np
import matplotlib.pyplot as plt

# predict function
def predict(x, a, b):
    return a*x + b

def compute_derivative(x, y_hat, y):
    da = 2*x*(y_hat-y)
    db = 2*(y_hat-y)
    return da, db

def update(a, b, lr, da, db):
    a = a - lr*da
    b = b - lr*db
    return a, b

def show_blackbox(a, b):
    x_min = -2
    x_max = 10
    x_values = np.linspace(x_min, x_max, 300)
    predict_fx = predict(x_values, a, b)
    
    # Visualization Stuff
    fig, ax = plt.subplots()
    ax.set_xlim([-2, 10])
    ax.set_ylim([-40, 40])
    ax.scatter([4], [6], zorder=4)
    line1, = ax.plot(x_values, predict_fx, 'g')
    ax.grid(True)
    

# given data 
x = 4
y = 6

a = -0.8
b = -4.0
lr = 0.01
```

```python
# show the current state
show_blackbox(a, b)
```

```python
# optimize a and b

for _ in range(10):
    y_hat = predict(x, a, b)
    da, db = compute_derivative(x, y_hat, y)
    a, b = update(a, b, lr, da, db)
```

```python
# after optimizing a and b
show_blackbox(a, b)
```
