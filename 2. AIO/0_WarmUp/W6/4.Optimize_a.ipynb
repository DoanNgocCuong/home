```python
import random
import numpy as np
import matplotlib.pyplot as plt

# predict function
def predict(x, a):
    return a*x

def compute_derivative(x, y_hat, y):
    return 2*x*(y_hat-y)

def update(a, lr, da):
    return a - lr*da

def show_blackbox(a):
    x_min = -2
    x_max = 10
    x_values = np.linspace(x_min, x_max, 300)
    predict_fx = predict(x_values, a)
    
    # Visualization Stuff
    fig, ax = plt.subplots()
    ax.set_xlim([-2, 10])
    ax.set_ylim([-40, 40])
    ax.scatter([4], [6], zorder=4)
    line1, = ax.plot(x_values, predict_fx, 'g')
    ax.grid(True)



# given data 
x = 4
y = 6

a = -6.0
lr = 0.01
```

```python
# show the current state
show_blackbox(a)
```

```python
# optimize a

for _ in range(10):
    y_hat = predict(x, a)
    da = compute_derivative(x, y_hat, y)
    a = update(a, lr, da)
```

```python
# after optimizing a
show_blackbox(a)
```
